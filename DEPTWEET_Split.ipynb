### 四分类

```python
import pandas as pd
import numpy as np

data = pd.read_excel("/data/zhengtianlong/Private/Depressed/deptweet_dataset.xlsx")
data.head()
```

```python
data = data.rename(columns = {"tweet":"text"})
data = data[['id', 'text', 'target', 'label']]
data.head()
```

```python
data.to_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet.csv', index=False)
data = pd.read_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet.csv')
data.head(5)
```

```python
data_0 = data[data['target'] == 0]
data_1 = data[data['target'] == 1]
data_2 = data[data['target'] == 2]
data_3 = data[data['target'] == 3]

print(len(data)==(len(data_0)+len(data_1)+len(data_2)+len(data_3)))
```

```python
from sklearn.model_selection import train_test_split
def train_validate_test_split(df, seed = 7):
    train, valtest = train_test_split(df, test_size=0.1, random_state = seed)
    val, test = train_test_split(valtest, test_size = 0.35, random_state = seed)
    return train, val, test

train_df_0, valid_df_0, test_df_0 = train_validate_test_split(data_0)
train_df_1, valid_df_1, test_df_1 = train_validate_test_split(data_1)
train_df_2, valid_df_2, test_df_2 = train_validate_test_split(data_2)
train_df_3, valid_df_3, test_df_3 = train_validate_test_split(data_3)

train_df = pd.concat([train_df_0, train_df_1, train_df_2, train_df_3])
valid_df = pd.concat([valid_df_0, valid_df_1, valid_df_2, valid_df_3])
test_df = pd.concat([test_df_0, test_df_1, test_df_2, test_df_3])

print(len(data)==(len(train_df)+len(valid_df)+len(test_df)))

train_df.to_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet_train.csv', index=False)
valid_df.to_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet_valid.csv', index=False)
test_df.to_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet_test.csv', index=False)
```

### 二分类

```python
import pandas as pd

# 读取CSV文件
data = pd.read_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet_train.csv')

# 将target列中的2和3替换为1
data['target'] = data['target'].replace({2: 1, 3: 1})

# 将label列中的moderate和severe替换为mild
data['label'] = data['label'].replace({'moderate': 'mild', 'severe': 'mild'})

# 如果需要，可以将修改后的DataFrame保存回CSV
data.to_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet_train_binary.csv', index=False)
```

```python
data = pd.read_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet_test.csv')

data['target'] = data['target'].replace({2: 1, 3: 1})
data['label'] = data['label'].replace({'moderate': 'mild', 'severe': 'mild'})
data.to_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet_test_binary.csv', index=False)

data = pd.read_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet_valid.csv')

data['target'] = data['target'].replace({2: 1, 3: 1})
data['label'] = data['label'].replace({'moderate': 'mild', 'severe': 'mild'})
data.to_csv('/data/zhengtianlong/Private/Depressed/Deptweet/Deptweet_valid_binary.csv', index=False)
```
